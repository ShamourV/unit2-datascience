```python
import pandas as pd
df = pd.read_csv('Avengers.csv')
df.head()
```

#What was the disparity in budget in Marvel movies on their opening weekend?

```python
df.groupby(['film', 'budget'], as_index=False)['opening weekend ($m)'].sum()
```

```python
den = df.groupby(['film', 'year'])['opening weekend ($m)'].sum()
den = df.groupby(['film', 'year'])['budget'].sum()
den.plot(kind='bar',xlabel='film', ylabel='opening weekend ($m)',title='Movie')
```

What was the difference in the first opening weekend and the second weekend of the movie.

```python
den = df.groupby(['film', 'year'])['opening weekend ($m)'].sum()
den = df.groupby(['film', 'year'])['second weekend ($m)'].sum()
den.plot(kind='bar',xlabel='film', ylabel='opening weekend ($m)',title='Movie')
```
